```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```
df = pd.read_csv("Hotel Booking Demand Archive.zip.csv")
```

```
df.drop('company',axis=1,inplace=True)
```

```
df['agent']=df['agent'].fillna(0)
```

```
df['country']=df['country'].fillna('Unknown')
```

```
df['children']= df['children'].fillna(df['children'].median())
```

```
df.duplicated().sum()
```

```
df.drop_duplicates(inplace=True)
```

```
df.duplicated().sum()
```

```
df[(df['adults']==0)&(df['children']==0)&(df['babies']==0)]
```

```
df= df[~((df['adults']==0)&(df['children']==0)&(df['babies']==0))]
```

```
df['total_guests']=df['adults']+df['children']+df['babies']
```

```
df[['adults','children','babies','total_guests']].head()
```

```
df['total_nights'] = (
    df['stays_in_weekend_nights'] +
    df['stays_in_week_nights']
)
```

```
df['revenue'] = (
    df['adr'] *
    df['total_nights']
)
```

```
month_order = [
'January','February','March','April',
'May','June','July','August',
'September','October','November','December'
]

df['arrival_date_month'] = pd.Categorical(
    df['arrival_date_month'],
    categories=month_order,
    ordered=True
)
```

```
df['adr'].describe()
```

```
df[df['adr'] > 1000]
```

```
df = df[df['adr'] < 1000]
```

```
df['booking_status'] = df['is_cancelled'].map(
    {
        0:'Not Cancelled',
        1:'Cancelled'
    }
)
```

```
df['lead_time_category'] = pd.cut(
    df['lead_time'],
    bins=[0,30,90,180,365,800],
    labels=[
        '0-30 Days',
        '31-90 Days',
        '91-180 Days',
        '181-365 Days',
        '365+ Days'
    ]
)
```

```
df.to_csv(
    "hotel_bookings_cleaned.csv",
    index=False
)
```

```
df['repeat_guest'] = df['is_repeated_guest'].map(
    {
        0:'New Customer',
        1:'Repeated Customer'
    }
)
```

```
df['customer_group'] = np.where(
    (df['children'] + df['babies']) > 0,
    'Family',
    'Non-Family'
)
```

```
monthly_bookings = df.groupby('arrival_date_month').size()

monthly_bookings.plot(
    kind='line',
    figsize=(10,5),
    marker='o',
    title='Monthly Booking Trend'
)

plt.xlabel('Month')
plt.ylabel('Bookings')
plt.show()
```

```
cancel_hotel = pd.crosstab(
    df['hotel'],
    df['is_cancelled'],
    normalize='index'
)*100

cancel_hotel.plot(
    kind='bar',
    figsize=(8,5),
    title='Cancellation Rate by Hotel Type'
)

plt.show()
```

```
segment_revenue = df.groupby(
    'market_segment'
)['revenue'].sum().sort_values(ascending=False)

segment_revenue.plot(
    kind='barh',
    figsize=(10,5),
    title='Revenue by Market Segment'
)

plt.show()
```

```
sns.boxplot(
    x='is_cancelled',
    y='lead_time',
    data=df
)

plt.title('Lead Time vs Cancellation')
plt.show()
```

```
df['customer_type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(7,7)
)

plt.title('Customer Type Distribution')
plt.ylabel('')
plt.show()
```
